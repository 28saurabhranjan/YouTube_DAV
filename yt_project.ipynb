2.Which countries produce the highest number of top YouTube creators, and is there a correlation between the country of origin and subscriber counts?

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Country", y="subscribers", data=df, palette="viridis")
plt.title("Top YouTube Creators by Country")
plt.xlabel("Country")
plt.ylabel("Subscribers (10 Millions)")
plt.xticks(rotation=90)
plt.show()
```

No there is no correlation between subscribers count and country.

3. How does upload frequency impact subscriber growth and video views for these top creators?

```python
top50=df[:50]
# Calculate metrics
top50["Views per Upload"] = top50["video views"] / top50["uploads"]
top50["Growth Rate"] = top50["subscribers"] / top50["uploads"]

plt.figure(figsize=(10,10))
plt.subplot(2, 1, 1)
plt.bar(top50["Youtuber"], top50["Views per Upload"], color="skyblue")
plt.title("Views per Upload")
plt.ylabel("Views")
plt.xticks(rotation=90)

```

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 1, 2)
plt.bar(top50["Youtuber"], top50["Growth Rate"], color="salmon")
plt.title("Subscriber Growth Rate")
plt.xlabel("Channel")
plt.ylabel("Subscribers")
plt.xticks(rotation=90)
```

4. What types of content are most popular among top YouTube creators, and how do they vary by subscriber count?

```python
# Group by content category and calculate average subscribers
category_stats = df.groupby("category").agg({
    "subscribers": "mean",
    "Youtuber": "count"
}).reset_index()
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(category_stats["category"], category_stats["subscribers"], color="skyblue")
plt.title("Average Subscribers by Content Category")
plt.xlabel("Content Category")
plt.ylabel("Average Subscribers(10 Millions)")
plt.xticks(rotation=90)
plt.show()
```

8. Is there any correlation between video views and subscribers count?

```python
from scipy.stats import pearsonr
subs=df['subscribers']
views=df['video views']

correlation_coeffecient, _ = pearsonr(views, subs)
print(f"Correlation Coefficient: {correlation_coeffecient:.2f}")

```

Yes there is correlation of coefficient of 0.75 between subscribers and views count.

9. Is there any Correlation between earning and subscribers?

```python
# Correlation between earning and subscribers
earning=df['highest_yearly_earnings']

correlation_coeff, _ = pearsonr(earning, subs)
print(f"Correlation Coeffecient: {correlation_coeff:.2f}")
```
